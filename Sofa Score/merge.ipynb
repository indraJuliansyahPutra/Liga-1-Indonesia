```python
import pandas as pd
import os

# Folder input dan output
data_pemain_folder = "sementara-Data Pemain"
rating_pemain_folder = "sementara-rating"
output_folder = "sementara-gabungan"

os.makedirs(output_folder, exist_ok=True)

# Ambil semua nama file di folder data pemain
files = [f for f in os.listdir(data_pemain_folder) if f.endswith(".csv")]

# Loop setiap klub berdasarkan nama file
for file_name in files:
    try:
        nama_klub = file_name.replace("-PEMAIN.csv", "")

        data_pemain_path = os.path.join(data_pemain_folder, file_name)
        rating_pemain_path = os.path.join(rating_pemain_folder, f"{nama_klub}-rating.csv")

        if not os.path.exists(rating_pemain_path):
            print(f"⚠️  Rating file not found for: {nama_klub}, skip.")
            continue
        
        df_data_pemain = pd.read_csv(data_pemain_path)
        df_rating_pemain = pd.read_csv(rating_pemain_path)
        df_rating_pemain = df_rating_pemain.rename(columns={"Link": "Link Pemain"})
        
        # Gabungkan
        df_gabungan = pd.merge(df_data_pemain, df_rating_pemain, on="Link Pemain", how="left")
        
        # Simpan
        output_path = os.path.join(output_folder, f"{nama_klub}.csv")
        df_gabungan.to_csv(output_path, index=False)
    
    except Exception as e:
        print(f"Gagal proses {file_name}: {e}")
```
