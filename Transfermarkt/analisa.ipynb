```python
import pandas as pd
import os

folder_path = "Data Pemain"
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    print(file)
    data = pd.read_csv(file_path)

    nama_file = file.replace(".csv", "")
    
    data['Nama'] = data['Nama'].str.upper()
    data['Klub'] = data['Klub'].str.upper()
    data['Posisi'] = data['Posisi'].str.upper()
    data['Tinggi'] = data['Tinggi'].str.replace("m", "").str.replace(".", ",", regex=False)
    data["Usia"] = data["Tanggal Lahir"].str.extract(r"\((\d+)\)").astype(int)
    data["Tanggal Lahir"] = data["Tanggal Lahir"].str.replace(r"\s*\(\d+\)", "", regex=True)
    data['Kaki Dominan'] = data['Kaki Dominan'].str.upper()

    data['Nomor Punggung'] = data['Nomor Punggung'].replace("-", "0")

    def ke_miliar(nilai):
        nilai = nilai.replace('Rp', '').replace('.', '')  # Hilangkan Rp dan titik akhir
        if 'Mlyr' in nilai:
            angka = float(nilai.replace('Mlyr', '').replace(',', '.'))
            return angka  # sudah dalam miliar
        elif 'Jt' in nilai:
            angka = float(nilai.replace('Jt', '').replace(',', '.'))
            return angka / 1000  # ubah juta ke miliar
        else:
            return 0.0
        
    data['Nilai Pasar'] = data['Nilai Pasar'].apply(ke_miliar).round(2)

    data.to_csv(f'Data Bersih\{nama_file}.csv', index=False)
```
