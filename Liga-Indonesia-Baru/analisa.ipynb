```python
import pandas as pd
import os
import numpy as np

folder_path = "Data Pemain"
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    print(file)
    data = pd.read_csv(file_path)

    kolom_diisi_nol = ['Penyelamatan', 'Gol', 'Tembakan', 
                       'Tembakan Tepat Sasaran', 'Assist', 'Tekel', 
                       'Intersepsi', 'Sapuan', 'Kartu Kuning', 
                       'Kartu Merah', 'Pelanggaran', 'Offside']

    kolom_umpan = "Umpan Sukses"
    mask = data['Penampilan'] == 0
    data.loc[mask, kolom_diisi_nol] = 0
    data.loc[mask, kolom_umpan] = "0/0"

    nama_file = file.replace('_', ' ')
    nama_file = nama_file.replace('-PEMAIN.csv', "")
    nama_file = nama_file.title()

    data[['Umpan Biasa', 'Umpan Total']] = data['Umpan Sukses'].str.split('/', expand=True)
    data['Umpan Sukses'] = data['Umpan Biasa']
    data = data.drop(columns=['Umpan Biasa'])

    data['Umpan Sukses'] = data['Umpan Sukses'].astype(float)
    data['Umpan Total'] = data['Umpan Total'].astype(float)
    data['Akurasi Passing'] = np.where(
        data['Umpan Total'] == 0,
        0,
        data['Umpan Sukses'] / data['Umpan Total']
    )

    data['Akurasi Tembakan'] = np.where(
        data['Tembakan'] == 0,
        0,
        data['Tembakan Tepat Sasaran'] / data['Tembakan']
    )

    data.to_csv(f'Data Bersih/{nama_file}.csv', index=False)
```
